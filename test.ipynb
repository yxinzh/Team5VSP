```python
import pandas as pd
```

```python
products = pd.read_csv('products.csv')
products = products.iloc[:,1:]
products.columns
```

```python
#make value style consistent with training date

products.loc[
    products["FRAMESHAPE"].str.contains("RECTANGLE"),
    "FRAMESHAPE"
] = "Rectangle"

products.loc[
    products["FRAMESHAPE"].str.contains("CAT EYE"),
    "FRAMESHAPE"
] = "Cat-Eye"

products.loc[
    ~ products["MATERIALCODE1"].str.contains("METAL"),
    "MATERIALCODE1"
] = "Plastic"

products.loc[
    products["MATERIALCODE1"].str.contains("REC METAL-STAINLESS"),
    "MATERIALCODE1"
] = "Metal"
```

```python
products['SUNOPTICAL'] = products['SUNOPTICAL'].str.capitalize()
products['MATERIALCODE1'] = products['MATERIALCODE1'].str.capitalize()
products['FRAMECONSTRUCTION'] = products['FRAMECONSTRUCTION'].str.capitalize()
products['FRAMESHAPE'] = products['FRAMESHAPE'].str.capitalize()
```

```python
products['Date'] = products['RELEASEDATE'].str.split().str[-1]
products['Date'] = pd.to_datetime(products['Date'])
```

```python
trends = pd.read_csv('opticalsun_googletrends.csv')
trends['Date'] = pd.to_datetime(trends['Time'])
products = products.merge(trends[['Date','Glasses_3m_avg','Sunglasses_3m_avg']],on='Date',how='left')
```
